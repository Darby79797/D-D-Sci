```python
df = pd.read_csv(url,index_col=0)
```

Observations:
Charisma and success rate has noticable jumps at 8 and 10.

Consitution is vaguely linearly increasing; jump at 18.

Dexterity anticorrelates with success; from the genre of data, I expect this is because having a high value on one stat is anticorrelated with having a high value in the others, and that dexterity is mostly useless.

Intelligence > 7 gains most/all benefits of intelligence.

Strength spikes at 8, and rises again at 13-15, then at 19.

Wisdom < 8 looks bad, >12 ok, and >15 desirable.


I get 10 points to augment stats of (4,14,13,13,6,12); initial best guess would be +4 CHA, +3 WIS, +3 STR.

```python
def successRateGraphs(df):
    fig, axs = plt.subplots(2, 3)
    
    sub = df[['cha','result']]
    successRate = []
    for i in range(21):
        t = sub[sub.cha == i]
        t1 = t['result'].value_counts(normalize=True)
        t2 = t1.get(key='succeed')
        if t2==None:
            t2=0
        successRate.append(t2)
    axs[0, 0].bar(range(len(successRate)),successRate)
    axs[0, 0].set_title("CHA")
    
    sub = df[['con','result']]
    successRate = []
    for i in range(21):
        t = sub[sub.con == i]
        t1 = t['result'].value_counts(normalize=True)
        t2 = t1.get(key='succeed')
        if t2==None:
            t2=0
        successRate.append(t2)
    axs[0, 1].bar(range(len(successRate)),successRate)
    axs[0, 1].set_title("CON")
    
    sub = df[['dex','result']]
    successRate = []
    for i in range(21):
        t = sub[sub.dex == i]
        t1 = t['result'].value_counts(normalize=True)
        t2 = t1.get(key='succeed')
        if t2==None:
            t2=0
        successRate.append(t2)
    axs[0, 2].bar(range(len(successRate)),successRate)
    axs[0, 2].set_title("DEX")
    
    sub = df[['int','result']]
    successRate = []
    for i in range(21):
        t = sub[sub.int == i]
        t1 = t['result'].value_counts(normalize=True)
        t2 = t1.get(key='succeed')
        if t2==None:
            t2=0
        successRate.append(t2)
    axs[1, 0].bar(range(len(successRate)),successRate)
    axs[1, 0].set_title("INT")
    
    sub = df[['str','result']]
    successRate = []
    for i in range(21):
        t = sub[sub.str == i]
        t1 = t['result'].value_counts(normalize=True)
        t2 = t1.get(key='succeed')
        if t2==None:
            t2=0
        successRate.append(t2)
    axs[1, 1].bar(range(len(successRate)),successRate)
    axs[1, 1].set_title("STR")
    
    sub = df[['wis','result']]
    successRate = []
    for i in range(21):
        t = sub[sub.wis == i]
        t1 = t['result'].value_counts(normalize=True)
        t2 = t1.get(key='succeed')
        if t2==None:
            t2=0
        successRate.append(t2)
    axs[1, 2].bar(range(len(successRate)),successRate)
    axs[1, 2].set_title("WIS")


    for ax in axs.flat:
        ax.set(xlabel='Stat Value', ylabel='P(success)')

    for ax in axs.flat:
        ax.label_outer()
    
```

```python
successRateGraphs(df)
```

```python
medDex = df[df.dex>=10]
highDex = df[df.dex>=13]

medCon = df[df.con>=10]
highCon = df[df.con>=14]

medInt = df[df.int>=10]
highInt = df[df.int>=13]

medWis = df[df.wis>=10]
highWis = df[df.wis>=12]

highCD = highDex[highDex.con>=14]
highCDI = highCD[highCD.int>=13]
highAll = highCDI[highCDI.wis>=12]
```

```python
successRateGraphs(medDex)
successRateGraphs(highDex)

successRateGraphs(medCon)
successRateGraphs(highCon)

successRateGraphs(medInt)
successRateGraphs(highInt)

successRateGraphs(medWis)
successRateGraphs(highWis)
```

```python
successRateGraphs(highAll)
```

Nothing here contradicts previous guesses; final answer is +4 CHA, +3 STR, +3 WIS
